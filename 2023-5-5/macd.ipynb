```python
'''
i:当天
x:当天收盘价
y：指数移动平均价EMA
com：时间窗口（移动范围）大小，可以换算为衰减系数
'''
# 计算指数平滑移动平均线SMA
def my_MACD(data):
    # arr=data['Clsprc']
    # alpha = 1/(1+com)
    # y=(1-alpha)*my_MACD(data,com-1)+alpha*com
    import numpy as np
    EMA12 = data['Clsprc'].rolling(12).mean()
    EMA26 = data['Clsprc'].rolling(26).mean()
    DIF = EMA12 - EMA26
    DIF.fillna(0, inplace = True)
    DEA = np.zeros((len(DIF)))
    MACD = np.zeros((len(DIF)))
    for t in range(len(DIF)):
        if t == 0:
            DEA[t] = DIF[t]
        if t > 0:
            DEA[t] = (2 * DIF[t] + 8 * DEA[t - 1]) / 10
        MACD[t] = 2 * (DIF[t] - DEA[t])
    return DIF,DEA,MACD
# 计算指数平滑移动平均线EMA
def MACD(data): 
    import numpy as np 
    EMA12 = data['Clsprc'].ewm(12)
    EMA26 = data['Clsprc'].ewm(26)
    DIF = EMA12.obj- EMA26.obj
    DEA = np.zeros((len(DIF)))
    MACD=np.zeros((len(DIF)))
    for t in range(len(DIF)):
        if t==0:
             DEA[t]= DIF[t]
        if t>0:
             DEA[t]=(2*DIF[t]+8*DEA[t-1])/10
        MACD[t]=2*(DIF[t]-DEA[t])
    return MACD
```
